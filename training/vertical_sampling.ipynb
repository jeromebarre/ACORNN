```python
# Plot
plt.figure(figsize=(5, 8))
plt.plot(weights, pressures_hPa, marker='o', color='royalblue')
plt.yscale("log")
plt.gca().invert_yaxis()  # Pressure increases downward
plt.grid(True, which='both', linestyle='--', alpha=0.5)

plt.xlabel("Sampling Weight")
plt.ylabel("Pressure (hPa)")
plt.title("Vertical Sampling Weights (Log Pressure Scale)")
plt.tight_layout()
plt.show()
```
